## Task 3 - Create a correlation matrix heatmap (colored).

### Create a correlation heatmap using seaborn:

```python
# Calculating the correlation matrix
correlation_matrix = guns_violence_df.corr()

# Plotting the correlation matrix as a heatmap
plt.figure(figsize=(20, 10))
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

1. **Number of Guns Involved:**
   - **n_killed (-0.0282):** There is a small negative correlation between the number of guns involved and the number of people killed. This suggests that incidents with more guns involved do not necessarily result in more fatalities.
   - **n_injured (-0.038843):** Similarly, there is a small negative correlation with the number of people injured, indicating that more guns involved do not necessarily lead to more injuries.

2. **Number Killed and Injured (-0.125425):** Contrary to what might be expected, there is a slight negative correlation between the number of victims killed and injured. This suggests that incidents with higher fatalities do not always have a high number of injuries, and vice versa.

3. **Victim and Suspect Counts:**
   - **Victim Count vs. Male Victims (0.89837):** There is a high positive correlation between the overall victim count and the number of male victims. This indicates that incidents with a higher number of victims often involve more male victims.
   - **Victim Count vs. Female Victims (0.43182):** There is a moderate positive correlation with the number of female victims.
   - **Suspect Count vs. Male Suspects (0.876094):** Similarly, there is a high positive correlation between the suspect count and the number of male suspects.
   - **Suspect Count vs. Female Suspects (0.30444):** There is a moderate positive correlation with the number of female suspects.

4. **Yearly Trends:**
   - **Year vs. Victim Count (-0.111763):** There is a negative correlation between the year and the victim count, suggesting a possible decrease in the number of victims over the years.
   - **Year vs. Suspect Count (-0.015266):** There is a slight negative correlation with the suspect count, indicating a potential decrease in the number of suspects over time.

These correlations provide insights into the relationships between the involvement of guns, the number of victims and suspects, their gender, and trends over the years. However, it's important to remember that correlation does not imply causation, and these relationships could be influenced by various other factors

## Task 6 - Create a categorical plot and interpret the results.

### Categorical plot and interpret the results.

```python
# Selecting a subset of columns for histograms to avoid overcrowding
selected_columns = ['year', 'month', 'day', 'n_guns_involved', 'n_killed', 'n_injured', 
                    'victim_count', 'suspect_count']


# Creating histograms for the selected columns using seaborn
plt.figure(figsize=(15, 12))

for index, column in enumerate(selected_columns):
    plt.subplot(3, 3, index + 1)
    sns.histplot(guns_violence_df[column], kde=False, bins=15)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

```

1. **Year:** Displays the distribution of incidents over different years, useful for observing trends or changes in the frequency of incidents over time.

2. **Month:** Shows the monthly distribution of incidents, which can help identify any seasonal variations in the occurrence of gun violence.

3. **Day:** The distribution of incidents across different days of the month, potentially revealing any day-specific patterns.

4. **Number of Guns Involved (n_guns_involved):** Illustrates how frequently different numbers of guns are involved in incidents.

5. **Number Killed (n_killed):** This histogram shows the distribution of fatalities per incident, offering insights into the severity of these incidents.

6. **Number Injured (n_injured):** Similar to 'n_killed', this provides a view of how injuries are distributed across incidents.

7. **Victim Count (victim_count):** Shows the overall distribution of the total number of victims (including both injured and killed) per incident.

8. **Suspect Count (suspect_count):** The distribution of the number of suspects involved in the incidents.

## Task 7 - Revisit the questions you generated in the previous task and write answers to those you can based on the exploration you’ve conducted so far. Add any new questions that may have arisen based on the early findings in your visual exploration.

```python
# 1. Trend Analysis: Gun Violence Incidents Over the Years
plt.figure(figsize=(10, 6))
sns.countplot(data=guns_violence_df, x='year', palette='Blues')
plt.xticks(rotation=45)
plt.title('Gun Violence Incidents Over the Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()

# 2. Trend Analysis: Gun Violence Incidents by Month
plt.figure(figsize=(10, 6))
sns.countplot(data=guns_violence_df, x='month', palette='Blues')
plt.title('Gun Violence Incidents by Month')
plt.xlabel('Month')
plt.ylabel('Count')
plt.show()

# 3. Geographical Analysis: Number of Gun Violence Incidents by State
plt.figure(figsize=(15, 6))
sns.countplot(data=guns_violence_df, x='state', order=guns_violence_df['state'].value_counts().index, palette='Blues')
plt.xticks(rotation=90)
plt.title('Number of Gun Violence Incidents by State')
plt.xlabel('State')
plt.ylabel('Count')
plt.show()

# 4. Victim and Incident Analysis: Distribution of the Number of Victims per Incident
plt.figure(figsize=(10, 6))
sns.histplot(data=guns_violence_df, x='victim_count', bins=30, kde=True, color='blue')
plt.title('Distribution of the Number of Victims per Incident')
plt.xlabel('Number of Victims')
plt.ylabel('Frequency')
plt.xlim(0, 50) # Limiting x-axis for better visualization
plt.show()

# Recalculating the age group totals for victims and suspects, as the code execution state was reset
age_group_victim_columns = ['victim_0-17', 'victim_18-30', 'victim_31-45', 'victim_46-60', 'victim_61+']
age_group_victim_totals = guns_violence_df[age_group_victim_columns].sum()

age_group_suspect_columns = ['suspect_0-17', 'suspect_18-30', 'suspect_31-45', 'suspect_46-60', 'suspect_61+']
age_group_suspect_totals = guns_violence_df[age_group_suspect_columns].sum()

# 5. Demographic Analysis: Age Profile of Victims in Gun Violence Incidents
plt.figure(figsize=(10, 6))
sns.barplot(x=age_group_victim_totals.index, y=age_group_victim_totals.values, palette='Blues')
plt.title('Age Profile of Victims in Gun Violence Incidents')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()

# 6. Demographic Analysis: Age and Gender Distribution of Suspects
plt.figure(figsize=(10, 6))
sns.barplot(x=age_group_suspect_totals.index, y=age_group_suspect_totals.values, palette='Blues')
plt.title('Age Profile of Suspects in Gun Violence Incidents')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

## Task 8 - Define any hypotheses that you can at this point. You’ll be testing these later on.

1. **Yearly Trend Hypothesis:**
      - H0 (Null Hypothesis): There is no significant change in the frequency of gun violence incidents over the years.
      - H1 (Alternative Hypothesis): There is a significant change in the frequency of gun violence incidents over the years.

2. **Seasonal Variation Hypothesis:**
      - H0: There is no significant seasonal or monthly variation in the occurrence of gun violence incidents.
      - H1: There is a significant seasonal or monthly variation in the occurrence of gun violence incidents.

3. **Geographical Distribution Hypothesis:**
      - H0: Gun violence incidents are evenly distributed across all states.
      - H1: Some states have significantly higher rates of gun violence incidents than others.

4. **Victim Count Variability Hypothesis:**
      - H0: The number of victims per incident is consistent across all gun violence incidents.
      - H1: There is significant variability in the number of victims per gun violence incident.

5. **Demographic Profile of Victims Hypothesis:**
      - H0: All age groups are equally likely to be victims in gun violence incidents.
      - H1: Certain age groups are more likely to be victims in gun violence incidents.

6. **Demographic Profile of Suspects Hypothesis:**
      - H0: All age groups are equally likely to be suspects in gun violence incidents.
      - H1: Certain age groups are more likely to be suspects in gun violence incidents.

7. **Correlation Between Victims/Suspects and Demographics Hypothesis:**
      - H0: There is no correlation between the number of victims/suspects and their demographic characteristics (age, gender).
      - H1: There is a significant correlation between the number of victims/suspects and their demographic characteristics (age, gender).

### Data Cleaning

```python
# Removing columns with high percentage of missing values and participant_name for privacy concerns
columns_to_remove = ['participant_relationship', 'location_description', 'participant_name']
cleaned_gun_violence_data = gun_violence_data.drop(columns=columns_to_remove)
```

```python
# Checking the shape of the dataset after removal
cleaned_dataset_shape = cleaned_gun_violence_data.shape

cleaned_dataset_shape
```

```python
# Further removing 'sources' and 'source_url' columns as they are deemed irrelevant for the analysis
additional_columns_to_remove = ['sources', 'source_url']
cleaned_gun_violence_data = cleaned_gun_violence_data.drop(columns=additional_columns_to_remove)

# Checking the shape of the dataset after the additional removal
cleaned_gun_violence_data.shape
```

```python
# Removing the 'incident_url_fields_missing' and 'incident_url' columns

columns_to_remove = ['incident_url_fields_missing', 'incident_url']

cleaned_gun_violence_data = cleaned_gun_violence_data.drop(columns=columns_to_remove)



# Checking the shape of the dataset after the removal

cleaned_gun_violence_data.shape
```

```python
# Removing the 'notes'

columns_to_remove = ['notes']

cleaned_gun_violence_data = cleaned_gun_violence_data.drop(columns=columns_to_remove)



# Checking the shape of the dataset after the removal

cleaned_gun_violence_data.shape
```

I created 2 new columns "victim_count"  and "suspect_count"

```python
def count_victims_and_suspects(participant_type):
    if pd.isna(participant_type):
        return 0, 0  # Return 0, 0 if the data is missing

    victim_count = participant_type.count("Victim")
    suspect_count = participant_type.count("Suspect")
    
    return victim_count, suspect_count

# Apply the function to each row in the DataFrame
cleaned_gun_violence_data[['victim_count', 'suspect_count']] = cleaned_gun_violence_data['participant_type'].apply(lambda x: count_victims_and_suspects(x)).apply(pd.Series)



```

```python
# Display few rows to verify the new columns

cleaned_gun_violence_data
```

```python
# Creating a new column 'participant_number'
cleaned_gun_violence_data['participant_number'] = cleaned_gun_violence_data['victim_count'] + cleaned_gun_violence_data['suspect_count']
```
